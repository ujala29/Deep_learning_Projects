```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical
```

```python
# Load the dataset
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
# mnist.load_data() returns two tuples: (X_train, y_train) and (X_test, y_test).

# X_train → Training images (60000 samples, 28x28 pixels).

# y_train → Training labels (60000 integers from 0–9).

# X_test → Test images (10000 samples).

# y_test → Test labels.
```

```python
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
X_train = X_train / 255.0
X_test = X_test / 255.0
# did normalization
```

```python
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
# one hot encoding[0,0,0,1,0,0,0]
```

```python
model = Sequential([
    Flatten(input_shape=(28,28)),  # Converts 28x28 image to 784 vector
    Dense(128, activation='relu'), # Fully connected layer with 128 neurons
    Dense(10, activation='softmax') # Output layer for 10 classes
])
# batch_size = number of samples the model sees before it updates the weights.
```

```python
model.compile(
    optimizer='adam',         # Algorithm to update weights
    loss='categorical_crossentropy', # Loss function for multi-class classification
    metrics=['accuracy']      # Metric to monitor
)
#
```

```python
model.fit(X_train, y_train, epochs=5, batch_size=32, validation_split=0.2)
# batch_size = number of samples the model sees before it updates the weights.
```

```python
test_loss, test_acc = model.evaluate(X_test, y_test)
print("Test Accuracy:", test_acc)
```

```python
predictions = model.predict(X_test)
print(predictions[0])       # Probabilities for first image
print(predictions[0].argmax()) # Predicted digit
print("actual",y_test[0].argmax())
```
